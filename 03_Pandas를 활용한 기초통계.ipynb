```python
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer = load_breast_cancer()
breast_cancer
```

```python
breast_cancer.keys()
```

```python
print(breast_cancer.DESCR)
```

```python
breast_cancer.feature_names.shape
```

```python
# 1. data 추출
x_data = breast_cancer['data']
y_data = breast_cancer['target']
feature_names = breast_cancer['feature_names']
target_names = breast_cancer['target_names']

print(x_data.shape, y_data.shape)
print(feature_names.shape, target_names.shape)
```

```python
target_names
```

```python
import pandas as pd
# 2. 추출한 data를 가지고 dataframe을 만든다.
feature_df = pd.DataFrame(x_data, columns=feature_names)
feature_df.head()
```

```python
feature_df.shape
```

```python
target_df = pd.DataFrame(y_data, columns=['target'])
target_df.head()
```

```python
feature = ['mean radius',	'mean texture',	'mean perimeter',	'mean area']
data_df = pd.concat([feature_df[feature], target_df], axis=1)
data_df.head()
```
