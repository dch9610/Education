#### 색 및 스타일 지정

- 색 : red/blue/green/yellow = r/b/g/y, 검정/흰색 = k/w, 청록/자홍 = c/m
- 선 스타일 : 실선 = -, dashed line = --, 점선 = :, dash-dot line = -.
- 선 두께 : linewidth= 숫자로 조정
- 마커(점) 스타일 : 원 = o(알파벳), 세모 = ^/v(알파벳)/>/<, 네모 = s(square), 별=*, 십자=+, 다이아=D(알파벳), x = x
- 마커(점) 크기 : s= 숫자로 조정

#### 라벨링

- legend : 0 (best;최적위치 자동선정) , 1(우측상단), 2(좌측상단), 3(좌측하단), 4(우측하단)

```python
size = [30, 40, 50]
price = [1.5, 2.4, 4.6]
plt.scatter(size, price, c='g')
plt.xlabel('size')
plt.ylabel('price')
plt.title('price-size')
plt.legend('data_1', loc='best') #loc = 'lower right'
plt.show()
```

- 그래프 안에 격자를 넣고 싶다면?

```python
year = [i for i in range(2010, 2020)]
price = [i for i in range(1, 11)]
plt.scatter(year, price, c='k', marker='^', s=50)
plt.xlabel('year')
plt.ylabel('price')
plt.title('price-year')
plt.legend('y', loc=0)
plt.grid()
plt.show()
```

- 하나의 그래프 안에 여러개의 plot을 넣고 싶다면?

```python
year = [i for i in range(2010, 2020)]
price_1 = [i for i in range(1, 11)]
price_2 = [i for i in range(11, 1, -1)]

plt.scatter(year, price_1, c='k', marker='^', s=50, alpha=0.5) #alpha = 불투명도
plt.plot(year, price_2, c='r', linestyle=':', linewidth=1, marker='o')
plt.xlabel('year')
plt.ylabel('price')
plt.title('price-year')
plt.legend('y', loc=0)
plt.grid()
plt.show()
```

```python
city = ['seoul', 'daejeon', 'daegu', 'busan']

#위도(lat), 경도(lon)
lat = [37, 36, 35, 35]
lon = [127, 128, 129, 130]

#인구밀도
pop_den = [16154, 2839, 2790, 4454]
```

Quiz) 위도와 경도에 맞춰서 도시를 scatter를 이용해 점으로 표현하되, 인구밀도에 비례하여 크기를 다르게 표현해보세요.

```python
for x, y, d in zip(lon, lat, pop_den):
    plt.scatter(x, y, s=d)
```

- city 이름도 함께 표현하고 싶다면? (text)

```python
plt.scatter(lon, lat, s=pop_den, c=['r', 'b', 'g', 'y'], alpha=0.5)
for x, y, c in zip(lon, lat, city):
    plt.text(x, y, c) # x좌표, y좌표, 글자
plt.show()
```

```python
for x, y, d, c in zip(lon, lat, pop_den, city):
    plt.scatter(x, y, s=d, alpha=0.5)
    plt.text(x, y, c)
```

#### pie(파이)

```python
label = ['a', 'b', 'c', 'd']
percent = [40, 30, 20, 10]
plt.pie(percent, labels=label, autopct='%.1f%%')
plt.show()
```

- 그래프를 저장하고 싶다면?

```python
label = ['a', 'b', 'c', 'd']
percent = [40, 30, 20, 10]
plt.pie(percent, labels=label, autopct='%.1f%%')
plt.savefig('graph.png')
plt.show()
```

- 그래프의 크기를 조절하고 싶다면?

```python
label = ['a', 'b', 'c', 'd']
percent = [40, 30, 20, 10]
plt.figure(figsize=(3,3))
plt.pie(percent, labels=label, autopct='%.1f%%')
plt.savefig('graph.png')
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
# 1. data 추출
x_data = breast_cancer['data']
y_data = breast_cancer['target']
feature_names = breast_cancer['feature_names']
target_names = breast_cancer['target_names']

print(x_data.shape, y_data.shape)
print(feature_names.shape, target_names.shape)

import pandas as pd
# 2. 추출한 data를 가지고 dataframe을 만든다.
feature_df = pd.DataFrame(x_data, columns=feature_names)
# target_name = target_names[0] + '/' + target_names[1]

target_df = pd.DataFrame(y_data, columns=['target'])
data_df = pd.concat([feature_df, target_df], axis=1)
data_df.head()
```

```python
sns.factorplot(x='target', y='mean radius', data=data_df, kind='box')
sns.factorplot(x='target', y='mean radius', data=data_df, kind='violin')
sns.factorplot(x='target', y='mean radius', data=data_df, kind='strip')
sns.factorplot(x='target', y='mean radius', data=data_df, kind='swarm')
plt.show()
```

```python
data_df.skew()
```

```python
plt.figure(figsize=(16, 10))
plt.subplot(211)
# sns.distplot(petition_data.votes)
sns.kdeplot(data=data_df['area error'], shade=True, legend=False)
plt.title('Distribution of area error')

plt.subplot(223)
# sns.distplot(petition_data.votes)
sns.kdeplot(data=data_df['area error'], shade=True, legend=False, color='r')
plt.title('Distribution of area error (0 - 100)')
plt.xlim(0, 100)

plt.subplot(224)
# sns.distplot(petition_data.votes)
sns.kdeplot(data=data_df['area error'], shade=True, legend=False, color='r')
plt.title('Distribution of area error (500 - 600)')
plt.xlim(500, )
plt.ylim(0, 0.001)
plt.show()
```
